# 3D Plotting in Matplotlib 
## 3D plots are enabled by importing the mplot3d toolkit, included in the main Matplotlib installation
### Example 1:

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection="3d") # creates 3-dimensional axes 
# Running code here will create 3D axes
t = np.linspace(0,4*np.pi,100);
x = np.cos(t)
y = np.sin(t)
z = t
ax. plot3D(x,y,z,'r')
ax.scatter3D(x,y,z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.grid([])
```

## Contour Plot - 1
### Example 2:


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(10,8))
x = np.linspace(-5,5,50)
y = np.linspace(-5,5,50)
X, Y = np.meshgrid(x,y)
Z = X**2 + Y**2
#plt.contour(X,Y,Z,10, colors='blue', linewidths = 2)
plt.contourf(X,Y,Z,10, cmap='jet') # Filled contour plot
plt.xlabel('x')
plt.ylabel('y')
```

##  Surface plot
### Example 3:

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection="3d") # creates 3-dimensional axes 

x = np.linspace(-5,5,50)
y = np.linspace(-5,5,50)
X, Y = np.meshgrid(x,y)
Z = X**2 + Y**2

ax. plot_surface(X,Y,Z,cmap='plasma')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Surface')
ax.grid([])
```

## Contour Plot-2
### Example 4:

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection="3d") # creates 3-dimensional axes 

x = np.linspace(-5,5,50)
y = np.linspace(-5,5,50)
X, Y = np.meshgrid(x,y)
Z = X**2 + Y**2
ax.view_init(60,35) # View angle
ax.contour3D(X,Y,Z,50,cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.grid([])
```

## Wireframes
