## Quiver Plot: Plot of Vector Field
### Example1: Plot the quiver plot of the vector field $\vec{F}  = -x ~\vec{i} + y ~\vec{j}$

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.figure(figsize=(8,8))
x = np.linspace(-5,5,10)
y = np.linspace(-5,5,10)
X, Y = np.meshgrid(x,y) # Mesh points
Fx, Fy = -X, Y
plt.quiver(X,Y,Fx, Fy,color='red',scale=50)
```

### Plot the quiver plot of the vector field  $\vec{v}=\sin x\cos y ~\vec{i} + \cos x\sin y ~\vec{j}$

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x = np.arange(0,2*np.pi+2*np.pi/20,2*np.pi/20)
y = np.arange(0,2*np.pi+2*np.pi/20,2*np.pi/20)

X,Y = np.meshgrid(x,y)

U = np.sin(X)*np.cos(Y)
V = -np.cos(X)*np.sin(Y)

fig, ax = plt.subplots(figsize=(8,8))

ax.quiver(X,Y,U,V,color='green', alpha=0.5, scale=15.0)

ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.axis([0,2*np.pi,0,2*np.pi])
```

## Gradient color quiver plot
### Plot the quiver plot of the vector field  $\vec{v}=\sin x\cos y ~\vec{i} + \cos x\sin y ~\vec{j}$

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x = np.arange(0,2*np.pi+2*np.pi/20,2*np.pi/20)
y = np.arange(0,2*np.pi+2*np.pi/20,2*np.pi/20)

X,Y = np.meshgrid(x,y)

U = np.sin(X)*np.cos(Y)
V = -np.cos(X)*np.sin(Y)


n = 0
color_array = np.sqrt(((U-n)/2)**2 + ((V-n)/2)**2)

fig, ax = plt.subplots(figsize=(8,8))
ax.quiver(X,Y,U,V,color_array)

ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_aspect('equal')
```

### Plot the quiver plot of the gradient of the scalar funtion $f = xe^{-x^2-y^2}$

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(-2,2.2,0.2)
y = np.arange(-2,2.2,0.2)

X, Y = np.meshgrid(x, y)
Z = X*np.exp(-X**2 -Y**2)
dx, dy = np.gradient(Z)

n = 0
color_array = np.sqrt(((dx-n)/2)**2 + ((dy-n)/2)**2)

fig, ax = plt.subplots(1,1,figsize=(8,8))
ax.quiver(X,Y,dx,dy,color_array)

ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_aspect('equal')
```
